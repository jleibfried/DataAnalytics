## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name

  * Total School Budget

```python
school_data_complete.columns
```

```python
# What I want to do is a for loop which makes a DataFrame for a specific school 
listOfSchools = school_data_complete.school_name.unique()
count = 0

for scho in listOfSchools:
    # print(school_0)
    # old code keeping it for cleverness and posterity
    school_0 = school_data_complete.loc[school_data_complete['school_name']==listOfSchools[count],:]
    
    # School Name
    schName = scho
    
    # School Type
    schType = school_0.type.unique()[0]
    
    # Total Students
    schStu = school_0['student_name'].count()
    
    # Total School Budget
    schBud = school_0.budget.unique()[0]
    
    # Per Student Budget
    perStuBud = schBud/schStu

    # Average Math Score
    avgSchMath = "{:.2f}".format(school_0["math_score"].mean())

    # Average Reading Score
    avgSchRead = "{:.2f}".format(school_0["reading_score"].mean())

    # % Passing Math
    passingSchMath = school_0.loc[school_data_complete["math_score"] >= 70].count()
    pSM = "{:.2f}%".format(passingSchMath[0]/schStu*100)

    # % Passing Reading
    passingSchRead = school_0.loc[school_data_complete["math_score"] >= 70].count()
    pSR = "{:.2f}%".format(passingSchRead[0]/schStu*100)

    # % Overall Passing (The percentage of students that passed math and reading.)
    passingSchReadMath = school_0.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].count()
    pSRM = "{:.2f}%".format((passingSchReadMath[0]/schStu*100))
    
    tempSchool = pd.DataFrame({"School Name": [scho], "School Type": [schType],"# of Students": [schStu], "Budget (in $)": [schBud],"$ per Student": [perStuBud], "Avg Math Score":[avgSchMath],"Avg Reading Score":[avgSchRead], "% Passing Math": [pSM] ,"% Passing Reading": [pSR],"% Passing both Math and Reading":[pSRM]})

    if count == 0:
        updateSchools = tempSchool
    else:
        updateSchools = pd.merge(updateSchools,tempSchool, how="outer")
    
    count = count + 1
    
updateSchools
```
